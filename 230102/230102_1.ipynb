# null 값 처리
- isnull() : null이면 True, null아니면 False를 리턴

```python
df_last.isnull()
```

```python
df_last.isnull().sum() # 칼럼별 null의 갯수 / 277개
```

```python
df_last.head()
```

```python
df_last['분양가격(㎡)'].isnull().sum()
```

```python
df_last['분양가격(㎡)'] = pd.to_numeric(df_last['분양가격(㎡)'], errors='coerce')
# errors='coerce' : error가 발생하는 데이터를 nan값으로 치환하고 데이터 타입변경
```

```python
df_last.groupby('지역명')['분양가격(㎡)'].mean()
```

```python
df_last['평당분양가격'] = df_last['분양가격(㎡)'] * 3.3
df_last.head()
```

```python
df_last['규모구분'].unique()
```

```python
# 데이터값 심플화
df_last['규모구분'] = df_last['규모구분'].str.replace('초과','')
df_last['규모구분'] = df_last['규모구분'].str.replace('이하','')
df_last['규모구분'] = df_last['규모구분'].str.replace(' ','')
df_last['규모구분'] = df_last['규모구분'].str.replace('전용','')
df_last['규모구분']
```

```python
df_last.head()
```

```python
df_last = df_last.drop(['분양가격(㎡)'], axis=1)
```

```python
df_last.head()
```

```python
df_last.groupby('지역명')['평당분양가격'].mean().plot(kind='bar')
plt.show()
```

```python
df_last.groupby('연도')['평당분양가격'].mean().plot(kind='bar')
plt.show()
```
