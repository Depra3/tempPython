1. 웹서버가 web script(html)를 return으로 받아옴
2. web script를 code를 통해서 필요한 내용만 추출 => beautifulsoup 이용

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 웹서버에 요청해서 web script를 받아옴
```

```python
url = 'https://movie.naver.com/movie/point/af/list.naver'
res = requests.get(url)
```

```python
res.text
```

```python
# web script를 beautifulsoup으로 parsing
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
#html.parser는 하나의 종류, 어떤 걸 사용하든 큰 차이는 없음
```

```python
soup
```

```python
# 상위 경로
path = 'table.list_netizen tbody > tr'
trs = soup.select(path)
# len(trs)
```

```python
# 영화 제목
path = 'td.title > a.movie.color_b'
for tr in trs:
    print(tr.select(path)[0].text)#[0] 인덱싱으로 태그 벗김
```

```python
# 영화 리뷰내용
path = 'td.title'

for tr in trs:
    print(tr.select(path)[0].text.split('\n')[5])
    print()
```

```python
# 작성자
path = 'td > a.author'

for tr in trs:
    print(tr.select(path)[0].text)
    print()
```

```python

for tr in trs:
    title = tr.select('td.title > a.movie.color_b')[0].text
    content = tr.select('td.title')[0].text.split('\n')[5]
    author = tr.select('td > a.author')[0].text
    
    print('Title : ', title)
    print('Comment : ', content)
    print('Author : ', author)
    print()
```
