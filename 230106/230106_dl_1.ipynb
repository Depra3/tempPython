```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import pandas as pd
import numpy as np
```

```python
data_set = np.loadtxt('ref/dl_data/ThoraricSurgery.csv', delimiter=',')
data_set
```

```python
# x, y 분리 / [:] - 모든 노드
x = data_set[:, 0:17]
y = data_set[:, 17]
print(x)
print(y)
```

```python
# 뉴럴 네트워크 구성
# input_dim : 입력값의 차원
# 30 : hidden layer의 node 갯수
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) # Dense layer
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# loss - 오차 / binary_crossentropy - 0과 1을 이용하는 함수 
# adam - 
```

```python
# neural network 학습
# epcohs - 전체 데이터에 대해서 몇번 학습 시킬 것인가
# batch_size - 전체 데이터를 일정 단위로 묶어서 네트워크에 입력
model.fit(x, y, epochs=100, batch_size=10)
```
