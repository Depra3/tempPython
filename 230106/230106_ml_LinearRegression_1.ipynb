```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 지도학습 > 회귀분석
```

```python
# 1. 데이터 import
df = pd.read_csv('ref/data/USA_Housing.csv')
df.head()
```

```python
# 2. 생략
# 3. 데이터 전처리
df.info()
```

```python
# 4. target(y), input data(x) 분리
x = df.drop(['Price', 'Address'], axis=1)
y = df['Price']
```

```python
print(x)
print(y)
```

```python
# 5. train data - test data 분리
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# 6. 머신러닝 알고리즘 sklearn에서 import
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
# 7. 머신러닝 알고리즘에게 data와 y를 주면서 학습(train data)
lin.fit(x_train, y_train)
```

```python
# 8.
preds = lin.predict(x_test)
```
