```python
sql = """
select 
    c.customer_id, sum(p.amount), email
from 
    payment p join customer c
    on c.customer_id = p.customer_id
group by
    c.customer_id
order by sum(p.amount) desc
limit 5;
"""

pd.read_sql(sql, db)
```

```python
df_c = pd.read_sql('select * from customer;', db)
df_p = pd.read_sql('select * from payment;', db)

idx = pd.merge(df_c, df_p, on = 'customer_id').groupby('customer_id')['amount']\
.sum().sort_values(ascending=False).index[:5]

df_c[df_c.customer_id.isin(idx)][['customer_id', 'email']]
```
