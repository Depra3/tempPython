```python
sql = '''create table movie_comments (
title varchar(50),
user_id varchar(20),
date varchar(10),
score int,
comments varchar(500)
);'''

cursor.execute(sql)
```

```python
for i in range(1, 11):
    url = 'https://movie.naver.com/movie/point/af/list.naver?&page={}'.format(i)
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    path = '#old_content tr'
    trs = soup.select(path)
    
    for tr in trs[1:]:
        try: 
            title = tr.select('td > a')[0].text
            user_id = tr.select('td > a')[2].text
            date = tr.select('td')[2].text.split('*')[-1]
            score = tr.select('td.title')[0].text.split('10점 중')[1].split('\n')[0]
            comments = tr.select('td.title')[0].text.split('10점 중')[1].split('\n')[2]

            sql = "insert into movie_comments values ('{}', '{}', '{}', {}, '{}')".format(title, user_id, date,score, comments)

            cursor.execute(sql)
        except:
            pass
```
