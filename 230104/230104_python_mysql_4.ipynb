```python
for i in range(1,101):
    url = 'https://movie.naver.com/movie/point/af/list.naver?&page={}'.format(i)
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    path = 'table.list_netizen tbody > tr'
    trs = soup.select(path)

    for tr in trs[1:]:
        try:
            title = tr.select('a.movie.color_b')[0].text
            score = tr.select('em')[0].text
            comments = tr.select('.title')[0].text.split('\n')[5]
            id = tr.select('.author')[0].text
            date = tr.select('.num')[1].text.split('*')[4]

            sql = 'insert into movie_comments values("{}", "{}", {}, "{}", "{}")'.format(title, id, score, comments, date)
            cursor.execute(sql)
        except:
            pass
        
# try ~ except 가 없으면 '' 나 "" 때문에 오류 발생
```
