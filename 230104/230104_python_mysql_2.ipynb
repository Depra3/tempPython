```python
# film 테이블에서 설명에 'boring'이라는 텍스트가 포함되고,
# 렌탈 비용이 0.99인 영화제목, 설명, 렌탈비용을 출력
df = pd.read_sql('select * from film;',db)
df
```

```python
# sql 방식
sql="""
select title, description, rental_rate
from film
where description like '%boring%' and
    rental_rate=0.99
"""
pd.read_sql(sql,db)
```

```python
# pandas 방식 / sql : like = pandas : .str.contains / pandas는 대소문자 구분함
df[(df['description'].str.contains('Boring')) & (df['rental_rate']==0.99)][['title', 'description', 'rental_rate']]
```

```python
# film 테이블에서 등급으로 그룹을 묶어서 각 등급별 갯수, 평균 렌탈비용 출력
```

```python
# sql
sql="""
select rating, count(rating), avg(rental_rate)
from film
group by rating
"""
pd.read_sql(sql,db)
```

```python
# pandas
df.groupby('rating').agg({'rating':np.size, 'rental_rate':np.mean})
```

```python
#sql
sql = """
select c.customer_id, c.email, sum(p.amount)
from customer c, payment p
where c.customer_id = p.customer_id
group by c.customer_id
order by sum(p.amount) desc
;
"""
pd.read_sql(sql,db).head(5)
```

```python
# pandas
df_c = pd.read_sql('select * from customer;',db)
df_p = pd.read_sql('select * from payment;',db)
idx = pd.merge(df_c,df_p, on='customer_id').groupby('customer_id')['amount'].sum().sort_values(ascending=False).index[:5]

df_c[df_c.customer_id.isin(idx)][['customer_id','email']]
```
