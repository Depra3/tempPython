```python
# 매출이 가장 많은 영화 제목
# film, inventory, rental, payment
sql = """
select f.film_id, f.title, sum(p.amount)
from film f, rental r, payment p, inventory i
where
    i.film_id = f.film_id and
    i.inventory_id = r.inventory_id and
    p.rental_id = r.rental_id
group by f.film_id;
"""
df_join = pd.read_sql(sql,db)
df_join.head()
```
